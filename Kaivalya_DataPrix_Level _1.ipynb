```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
pd.set_option("display.max_columns", None)
df
```

```python
print("shape:",df.shape)
df.describe()
```

```python
df1 = df.drop("Subject Id",axis=1)
print("shape:",df1.shape)
for i in df1.isnull().sum():
    print(i)
```

```python
df1 = df1.fillna(value=df.median())
df1 = df1.fillna(method="bfill")
df1 = df1.fillna(method="ffill")
df1
```

```python
print("shape:",df1.shape)
sum(df1.isnull().sum())
```

```python
df1 = df1.drop("Metabolic Syndrome Type",axis=1)
df1
```

```python
df2 = df1.replace({
    'Gene A': {'Yes': 1, 'No': 0},
    'Gene B': {'Yes': 1, 'No': 0},
    'Gene C': {'Yes': 1, 'No': 0},
    'Gene D': {'Yes': 1, 'No': 0},
    'Breathing Rate': {'Tachypnea': 1, 'Normal (30-60)': 0},
    'Pulse rate': {'Tachycardia': 1, 'Normal': 0},
    'Gastrin Defect': {'Yes': 1, 'No': 0},
    'Neural Anomaly': {'Yes': 1, 'No': 0},
    'Assistance needed in fertility': {'Yes': 1, 'No': 0},
    'Previous maternal pregnancy record': {'Yes': 1, 'No': 0},
                  })
df2.dtypes
```

```python
df2.dtypes
```

```python
categorical_data = [
'Gender',
'Presence of severe allergies',
'Premature Delivery',
'CMP results',
]
```

```python
for i in categorical_data:
    print(df2[i].unique())
```

```python
dummies=pd.get_dummies(df2[categorical_data])
dummies
```

```python
droppers = ['Gender_Ambiguous',
 'Presence of severe allergies_-',
 'Premature Delivery_-',
 'CMP results_inconclusive']
```

```python
good_dummies = dummies.drop(droppers,axis=1)
good_dummies
```

```python
merged=pd.concat([df2,good_dummies],axis=1)
merged
```

```python
ONE=merged.drop(categorical_data,axis=1)
ONE
```

```python
sum(ONE.isnull().sum())
```

```python
y=df['Metabolic Syndrome Type']
y
```

```python
x=ONE
```

```python
print("Input variable x:", x.shape)
print("Target variable y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,train_size=0.8)
```

```python
print("Input variable x_train:", X_train.shape)
print("Target variable y_train:", y_train.shape)
print("Input variable x_test:", X_test.shape)
print("Target variable y_test:", y_test.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import f1_score
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
y_train_pred = xgb.predict(X_train)
y_test_pred = xgb.predict(X_test)

# Training set performance
xgb_train_accuracy = accuracy_score(y_train, y_train_pred) # Calculate Accuracy
xgb_train_mcc = matthews_corrcoef(y_train, y_train_pred) # Calculate MCC
xgb_train_f1 = f1_score(y_train, y_train_pred, average='weighted') # Calculate F1-score

# Test set performance
xgb_test_accuracy = accuracy_score(y_test, y_test_pred) # Calculate Accuracy
xgb_test_mcc = matthews_corrcoef(y_test, y_test_pred) # Calculate MCC
xgb_test_f1 = f1_score(y_test, y_test_pred, average='weighted') # Calculate F1-score

print('Model performance for Training set')
print('- Accuracy: %s' % xgb_train_accuracy)
print('- MCC: %s' % xgb_train_mcc)
print('- F1 score: %s' % xgb_train_f1)
print('----------------------------------')
print('Model performance for Test set')
print('- Accuracy: %s' % xgb_test_accuracy)
print('- MCC: %s' % xgb_test_mcc)
print('- F1 score: %s' % xgb_test_f1)
```

```python

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(3) # Define classifier
knn.fit(X_train, y_train) # Train model

# Make predictions
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)

# Training set performance
knn_train_accuracy = accuracy_score(y_train, y_train_pred) # Calculate Accuracy
knn_train_mcc = matthews_corrcoef(y_train, y_train_pred) # Calculate MCC
knn_train_f1 = f1_score(y_train, y_train_pred, average='weighted') # Calculate F1-score

# Test set performance
knn_test_accuracy = accuracy_score(y_test, y_test_pred) # Calculate Accuracy
knn_test_mcc = matthews_corrcoef(y_test, y_test_pred) # Calculate MCC
knn_test_f1 = f1_score(y_test, y_test_pred, average='weighted') # Calculate F1-score

print('Model performance for Training set')
print('- Accuracy: %s' % knn_train_accuracy)
print('- MCC: %s' % knn_train_mcc)
print('- F1 score: %s' % knn_train_f1)
print('----------------------------------')
print('Model performance for Test set')
print('- Accuracy: %s' % knn_test_accuracy)
print('- MCC: %s' % knn_test_mcc)
print('- F1 score: %s' % knn_test_f1)
```
